<h1> Difference between Linear and Polynomial Regression</h1>
<h2>Importing the Required Libraries</h2>
<ul>
    <li>matplotlib.pyplot</li>
    <li>numpy</li>
</ul>

```python
import matplotlib.pyplot as plt
import numpy as np
```

<b>Creating X & Y Variables of DataSet and then Plotting it.</b>

```python
np.random.seed(0)
x=2-3*np.random.normal(0,1,20)
y=x-2*(x**2)+0.5*(x**3)+np.random.normal(-3,3,20)
plt.scatter(x,y,s=20)
plt.show()
```

```python
x=x[:,np.newaxis]
y=y[:,np.newaxis]
```

<b>Creating Linear Regression Model</b>
    <br>
<b>And Plotting The obtained result</b>

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
plt.scatter(x,y,s=10)
plt.plot(x,y_pred,color='r')
plt.show()
```

<b>Importing The Library r2_score From sklearn.metrics</b>
    <br>
<b>Calculating The r2 Score For Linear Regression Model</b>

```python
from sklearn.metrics import r2_score
r2 =r2_score(y,y_pred)
print("r2_score: ",r2)
```

<b>Importing The Library PloynomialFeatures From sklearn.preprocessing</b>
    <br>
<b>Creating Polynomial Regression Model of degree 2</b>

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2)
x_poly=poly_features.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
y_poly_pred = model.predict(x_poly)
r2 =r2_score(y,y_poly_pred)
print("r2_score: ",r2)
```

<b>Plotting Graph For Polynomial Regression Model Of Degree 2</b>

```python
plt.scatter(x,y,s=10)
import operator
sort_axis=operator.itemgetter(0)
sorted_zip=sorted(zip(x,y_poly_pred),key=sort_axis)
x,y_poly_pred=zip(*sorted_zip)
plt.plot(x,y_poly_pred,color='r')
plt.show()
```

<b>Creating Polynomial Regression Model of degree 3</b>

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=3)
x_poly=poly_features.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
y_poly_pred = model.predict(x_poly)
r2 =r2_score(y,y_poly_pred)
print("r2_score: ",r2)
```

<b>Plotting Graph For Polynomial Regression Model Of Degree 3</b>

```python
plt.scatter(x,y,s=10)
import operator
sort_axis=operator.itemgetter(0)
sorted_zip=sorted(zip(x,y_poly_pred),key=sort_axis)
x,y_poly_pred=zip(*sorted_zip)
plt.plot(x,y_poly_pred,color='r')
plt.show()
```

<b>Creating Polynomial Regression Model of degree 4</b>

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=4)
x_poly=poly_features.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
y_poly_pred = model.predict(x_poly)
r2 =r2_score(y,y_poly_pred)
print("r2_score: ",r2)
```

<b>Plotting Graph For Polynomial Regression Model Of Degree 4</b>

```python
plt.scatter(x,y,s=10)
import operator
sort_axis=operator.itemgetter(0)
sorted_zip=sorted(zip(x,y_poly_pred),key=sort_axis)
x,y_poly_pred=zip(*sorted_zip)
plt.plot(x,y_poly_pred,color='r')
plt.show()
```

<b>Creating Polynomial Regression Model of degree 5</b>

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=5)
x_poly=poly_features.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
y_poly_pred = model.predict(x_poly)
r2 =r2_score(y,y_poly_pred)
print("r2_score: ",r2)
```

<b>Plotting Graph For Polynomial Regression Model Of Degree 5</b>

```python
plt.scatter(x,y,s=10)
import operator
sort_axis=operator.itemgetter(0)
sorted_zip=sorted(zip(x,y_poly_pred),key=sort_axis)
x,y_poly_pred=zip(*sorted_zip)
plt.plot(x,y_poly_pred,color='r')
plt.show()
```

When we compare all the reult we got. And see the r2 Score. We found out that among all the Polynomial Regression Model 2, 3, 4 and 5, the Poplynomial Regression Model of Degree 2 is found to be Best. Which says that it is best in terms of Accuracy.
